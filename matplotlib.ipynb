```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
t = np.linspace(-np.pi, np.pi, 256, endpoint=True)
c, s = np.cos(t), np.sin(t)
```

```python
plt.plot(t, s)
```

```python
plt.plot(t, c)
```

```python
plt.plot(t, s)
plt.plot(t, c)
```

```python
plt.plot(t, s, 'b-.')
plt.plot(t, c, 'o')
```

```python
plt.plot(t, s, color = 'red', linestyle= '-', linewidth = 5)
plt.plot(t, c, color = 'blue', linestyle ='-.', linewidth = 9)
plt.grid()
plt.xlim(-np.pi, np.pi)
```

```python
plt.subplot(2, 1, 1)
plt.plot(t, s)
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(t, c)
plt.grid()
```
